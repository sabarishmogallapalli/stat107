## Part 2.1: Create a KMeans Model

First, let's create a new KMeans model called `kmeans` that contains **just two (2) clusters**:

````python
# Import the KMeans library:
from sklearn.cluster import KMeans

# Create a KMeans model:
kmeans = KMeans(n_clusters = 2)
````

## Part 2.2: Fit the Numeric Columns to the Model

The KMeans model only works on numeric data *(you can't find a mean -- or average -- of non-numbers)*.  The following code provides a list of all the numeric columns.  Run this cell store the `numeric_columns` value:

````python
numeric_columns = ['At what age do you think you will die?', 'From 1 to 10, how happy are you right now?', 'How many classes are you taking this semester?', 'How many credit hours are you taking this semester?.1', 'How many classes do you have on Friday?', 'How many countries have you been to outside the US?', 'How many hours of sleep did you get last night?', 'How many hours of sleep do you get on average?', 'How many meals do you have one day?', 'How many pets do you have?', 'How many roommates do you have?', 'How many siblings do you have?', 'How many years do you plan to study at university?', ]
numeric_columns
````

Try to `fit` the `kmeans` model data from the `numeric_columns`. 

**Note**: You should expect to see a `ValueError: Input X contains NaN.`!  Make sure to get that error message.

````python
# Attempt to `fit` the `kmeans` model:
kmeans.fit(df[numeric_columns])
````

## Part 2.3: Find the Rows with Missing Numeric Data Columns

The error message above provides details on several solutions to dealing with missing data.  To continue to use `KMeans`, our best option is to drop rows with missing data.  Find the rows that have missing data in the `numeric_columns` and store those rows in `drops`:

````python
notNA = df[numeric_columns]
clean = notNA.dropna()
drops = df[~df.index.isin(clean.index)]
drops
````

## Part 2.4: Drop the Rows with Missing Data

Provided below is a snippet of code that creates a new DataFrame, `df2`, that contains only a copy of the rows that can be used for clustering:

````python
df2 = df[ ~df.index.isin(drops.index) ].copy()
df2
````

## Part 2.5: Normalize the Numeric Data

In lecture, you learned that means are **VERY** sensitive to outliers.  Since our data comes from different ranges, we must normalize the numeric data.  To do this, we can divide each column by the maximum value of the column, and this is done in the provided code below.

Notice that, in the result, all column values are now scaled to be between 0 and 1.  Since all values are now in the same range, we no longer have any column with values in a different range.

````python
for column in numeric_columns:
    df2[column] = df2[column] / df2[column].abs().max()
df2[numeric_columns]
````

## Part 2.6: Train the KMeans model with `numeric_columns` from `df2`

Finally, we can train our model!  Using the `numeric_columns` and `df2`, `fit` your `kmeans` model:

````python
kmeans.fit(df2[numeric_columns])
````

## Part 2.7: Find the KMeans Centroids


In lecture, you learned that `kmeans.cluster_centers_` will display the location of the final centroids.

The order they're displayed will be the same order as the columns are listed in `numeric_columns`.  Explore the centroids and numeric columns:

````python
kmeans.cluster_centers_
````

````python
numeric_columns
````

## Part 2.8: Predict the Cluster for Every Data Point

Using your `kmeans` model and the `numeric_columns`, `predict` the centroid for each row and store that prediction in a new column in `df2` called `cluster`:

````python
df2["cluster"] = kmeans.predict(df2[numeric_columns])
df2
````

## Part 2.9: Explore the Clusters Visually

In the code below, a scatter plot is used to display the two clusters in two different colors.  When using the columns `From 1 to 10, how happy are you right now?` and `How many siblings do you have?`, there are two very clean clusters divided by happiness.

````python
df2.plot.scatter(
    x="From 1 to 10, how happy are you right now?",
    y="How many siblings do you have?",
    c="cluster", colormap="Set1"
)
````

## DataFrame vs. Series

Before we continue, a brief lesson on some of the datatypes you'll encounter in this class. 

In `pandas`, there are three data structures we will be using, two of which you have already seen:

- A **DataFrame** is similar to an Excel sheet. It contains zero or more rows and zero or more columns. It can contain up to an unlimited number of rows and columns and is the basic data structure we've worked with from the start

- A **GroupBy** is a grouping of rows within a DataFrame. This "three dimensional data" requires us to use `agg` to reduce it back down to a DataFrame - which you've been doing in homeworks and labs thus far!

- A **Series** is a single row with zero or more values (columns). You can think of a DataFrame as a collection of **Series** objects. When we deal with a single row, we will usually want to deal with a **Series** instead of a **one row DataFrame**

In this lab, we will learn how to convert from a **one row DataFrame** to a **Series**, but here's a quick recap on how to convert from types:

### Converting Types

The `pandas` library makes it easy to convert between a **Series**, **DataFrame**, and **GroupBy**.

- From a **Series** (`row`) to a **DataFrame**
    - `row.to_frame()`
- From a **DataFrame** (`df`) to a **GroupBy**:
    - `df.groupby(by='columnName')` -- (You know this already!)
- From a **GroupBy** (`group`) to a **DataFrame**:
    - `df.agg( function )` -- (You know this already!)
- From a **DataFrame** (`df`) to a **Series**:
    - `df.iloc[0]`, if the DataFrame has one row

See this guide, https://discovery.cs.illinois.edu/guides/DataFrame-Fundamentals/dataframe-loc-vs-iloc/, for more information about `iloc[]`.

Now that you have chosen the **questions** you'd like to use to find your DISCOVERY twin, it's time to create a **function** to calculate a similarity value for each person, or each row in `df_hello`.

In lecture, remember learning that a Python **function** takes an input in and returns a **value**. For this specific function, we will take a **row** and calculate a **similarity score** for that row.

### Puzzle 2.1: The Similarity Function
We have provided the basic outline for the `similarity` function below. Follow the **commented details** and **complete the function** using all of your questions below:

````python
# we define the function "similarity" to take in a row, stored as the variable "row"

def similarity(row):
    # This function will keep track of a `score` of how similar two rows are with each other.
    # The score begins at zero and we will add to it every time a row matches.
    score = 0 

    # This if statement below checks to see if the `row` shares your choice between Pancakes and Waffles
    # - If the values are the same (==), we add one to the score.
    if row['Android or iOS?'] == my_row['Android or iOS?']: score = score + 1
    if row['Canes or Chick-fil-a?'] == my_row['Canes or Chick-fil-a?']: score = score + 1
    if row['Marvel or DC?'] == my_row['Marvel or DC?']: score = score + 1
    if row['UIUC Football or Basketball Team?'] == my_row['UIUC Football or Basketball Team?']: score = score + 1
    if row['In-State or Out-of-State/International'] == my_row['In-State or Out-of-State/International']: score = score + 1
    if row['Introvert or Extrovert?'] == my_row['Introvert or Extrovert?']: score = score + 1
    if row['Morning Person or Night Person?'] == my_row['Morning Person or Night Person?']: score = score + 1
    if row['Do you like partying at the University of Urbana-Champaign or do you want to go back home?'] == my_row['Do you like partying at the University of Urbana-Champaign or do you want to go back home?']: score = score + 1
    if row['Rap Music or Pop Music?'] == my_row['Rap Music or Pop Music?']: score = score + 1
    if row['Starbucks or Dunkin?'] == my_row['Starbucks or Dunkin?']: score = score + 1
    if row['Spring or Fall?'] == my_row['Spring or Fall?']: score = score + 1
    if row['TV Shows or Movies?'] == my_row['TV Shows or Movies?']: score = score + 1

    # Use this syntax for each question (column) that you have chosen in your list.
    # IMPORTANT: Make sure to keep your lines all at this same level of tabbing out from the start.    
    # Delete or the replace the "Pancakes or Waffles" question, unless you've decided that is one of your evaluating questions. 

    # Finally, return the `score`.
    # ...this line should always be the LAST line.
    return score
````

### Puzzle 2.2: Applying the Similarity Function
Now that you've properly defined the similarity function, we need to **apply** it to the rest of the Hello Dataset DataFrame so we can produce a **similarity score** for each person who took the survey!

Luckily, `pandas` provides a useful function for our use case: `df.apply()`.

Quick lesson: `df.apply(some_function)` applies a pre-defined function named `some_function` along an axis of your DataFrame. Because we want to run our `similarity` function on all **rows**, when you call `df.apply()`, ensure you use **`axis=1`**. This would look something like `df.apply(some_function, axis=1)`.

Using the above explanation as reference, **add a column** named `Similarity_Score` to your original DataFrame that contains the value after running your similarity function you defined earlier across all rows:

````python
df_hello['Similarity_Score'] = df_hello.apply(similarity,axis=1)
````

#### **The `contains` Function**
In Python, the `contains` function can be used to check if a specific **substring** is present in a **string** 

This will come in handy for us, because we can see if 'CA ', 'TA ', or 'Prof. ' is in the **names** of any rows in our DataFrame

Part of the statement has been completed for you to show you how to use `contains`. 

*Note that we have to cast our `Name` column of `df_hello` to a **string** by using `.str`

For more information, see https://discovery.cs.illinois.edu/guides/DataFrame-Row-Selection/dataframe-string-contains/

````python
df_staff = df_hello[(df_hello['Name'].str.contains('CA ')) + df_hello['Name'].str.contains('TA ') + df_hello['Name'].str.contains('Prof. ')] 
df_staff
# IMPORTANT: Make sure you include a whitespace in 'CA ' and 'TA ' so that people with those letters in their name aren't included. Whitespace counts. 
````
